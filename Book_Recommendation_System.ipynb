```python
book_new = book.pivot_table(index = 'UserId', columns = 'Title', values = 'Rating')
```

```python
#Imputing the NaN values with 0
book_new.fillna(0, inplace = True)
```

```python
book_new.index = book['UserId'].unique()
book_new
```

##Calculating Cosine Similarities between Users

```python
from sklearn.metrics import pairwise_distances
user1 = 1 - pairwise_distances(book_new.values, metric = 'cosine')
user1
```

```python
np.fill_diagonal(user1, 0)
```

```python
#Store the results in a dataframe
user = pd.DataFrame(user1)
user
```

```python
#Set the index and column names to user ids
user.index = book.UserId.unique()
user.columns = book.UserId.unique()
```

##Considering the most similar customer

```python
def recommend(custID):
  simID = user.loc[custID].idxmax()
  simID_book = book[book['UserId'] == simID].Title
  custID_book = book[book['UserId'] == custID].Title
  return set(simID_book) - set(custID_book)
```

```python
recommend(276813)
```

```python
recommend(507)
```

```python
recommend(3462)
```
